### Lets search for Hotels nearby

Hotels search using foursquare API

```python
# search for Cafeteria
search_query = 'Hotel'
radius = 10000

# Define the corresponding URL
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(ClIENT_ID, ClIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
url
```

Hotel Results

```python
hotel_data = requests.get(url).json()
venues = hotel_data['response']['venues']

# DataFrame
hotel_dataframe = json_normalize(venues)
hotel_dataframe.head()
```

### Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
hotel_columns = ['name', 'categories'] + [col for col in hotel_dataframe.columns if col.startswith('location.')]+ ['id']
hotel_filtered = hotel_dataframe.loc[:,hotel_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list4 = row['categories']
    except:
        categories_list4 = row['venue.categories']
        
    if len(categories_list4) == 0:
        return None
    else:
        return categories_list4[0]['name']

# filter the category for each row
hotel_filtered['categories'] = hotel_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
hotel_filtered.columns = [column.split('.')[-1] for column in hotel_filtered.columns]

hotel_filtered.head()
```

```python
hotelArray= ['Hotel']
hotel_filtered= hotel_filtered.loc[hotel_filtered['categories'].isin(hotelArray)]

# Retain Imp columns
hotel_filtered_final= hotel_filtered.drop(['labeledLatLngs', 'distance', 'cc', 'city', 'country', 'formattedAddress',\
                                        'crossStreet', 'id'], axis=1)
hotel_filtered_final.head()
```

```python
# Viewing the data in the data frame
hotel_filtered_final
```

```python
# delete rows with NaN values
hotel_filtered_final = hotel_filtered_final.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
hotel_filtered_final
```

```python
# define the dataframe columns
column_names = ['Name', 'Categories', 'Address', 'Latitude', 'Longitude' , 'State', 'Postal Code'] 

hotel = hotel_filtered_final.rename(columns = {"name": "Name", 
                                  "categories":"Categories", 
                                  "address": "Address",
                                  "lat":"Latitude",
                                  "lng":"Longitude",
                                  "state":"State",
                                  "postalCode":"PostalCode"})

hotel
```

```python
hotel_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around the Conrad Hotel

# add the Hotels as blue circle markers
for lat, lng, label in zip(hotel_filtered_final.lat, hotel_filtered_final.lng, hotel_filtered_final.categories):
    folium.features.Marker(
        [lat, lng],
        icon=folium.Icon(color='blue', icon='glyphicon glyphicon-home')
    ).add_to(hotel_map)

# display map
hotel_map
```
